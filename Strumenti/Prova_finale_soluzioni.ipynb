#### 3. Definisci un outlier

Un outlier è un dato anomalo molto distante dal valore mediamo. Solitamente vengono considerati outliers i dati al di sopra del terzo quartile più 1.5 della variazione interquartile (IQR) o al sotto del primo quartire meno 1.5 volte l'IQR
